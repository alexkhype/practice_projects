```python
fam = ['liz', 1.73, 'emma', 1.68, 'mom', 1.71, 'dad', 1.89]

# Cambia valor de elemento por medio de indice
fam[-1] = 1.86
print(fam)
```

```python
# Cambia valores por medio de slicing
fam[0:2] = ['lisa', 1.75]
print(fam)
```

```python
me = ['alex', 1.68]

# concatena una lista
ext_fam = fam + me
print(ext_fam)
```

```python
# elimina valor por medio de indice
del fam[2]
fam
```

```python
x = ['a','b','c']

# copia la referencia de x, no la lista
y = x
y[0] = 'z'

# X también cambió porque 'x' y 'y' sólo son referencias a las listas, no las listas en sí
x
```

```python
x = ['a','b','c']

# copia todos los elementos de x
y = x[:]

# copia x usando la función list()
z = list(x)

y[0] = 'z'
z[0] = 'w'

print(x,y,z,sep='---')
```

```python
# string to experiment with: place
place = "poolhouse"

# Use upper() on place
place_up = place.upper()

# Print out place and place_up
print(place)
print(place_up)

# Print out the number of o's in place
print(place.count("o")) 
```

```python
# Create list areas
areas = [11.25, 18.0, 20.0, 10.75, 9.50]

# Print out the index of the element 20.0
print(areas.index(20.0))

# Print out how often 9.50 appears in areas
print(areas.count(9.50))
```

```python
# Create list areas
areas = [11.25, 18.0, 20.0, 10.75, 9.50]

# Use append twice to add poolhouse and garage size
areas.append(24.5)
areas.append(15.45)

# Print out areas
print(areas)

# Reverse the orders of the elements in areas
areas.reverse()

# Print out areas
print(areas)
```
