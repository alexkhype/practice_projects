## While loop

```python
error = 50

while error > 1:
    error = error / 4
    print(error)
```

## In a dictionary

```python
world = {'afghanistan' : 30.55,
         'albania' : 2.77,
         'algeria' : 39.21}


# Necessary to use METHOD .items() for 
for key, value in world.items():
    print(f'{key}--{value}')
```

## In an array

```python
import numpy as np

height = np.array([1.73, 1.68, 1.71, 1.89, 1.79])
weight = np.array([65.4, 59.2, 63.8, 88.4, 68.7])
```

```python
# One-dimension array

bmi = weight / height ** 2
for val in bmi:
    print(val)
```

```python
# Two-dimension array

measures = np.array([height, weight])

# Aquí 'val' es toda el array
for val in measures:
    print(val)

# Para que haya iteración, se requiere la FUNCIÓN: np.nditer()
for val in np.nditer(measures):
    print(val)
```
