### Python list operations

#### Selecting Elements
- By index:

my_list[index] # e.g., my_list

- By slice:

my_list[start:end] # e.g., my_list[1:4]

- By condition (list comprehension):

[x for x in my_list if x > 10]

- Specific multiple indices:

[my_list[i] for i in ]


#### Updating Elements
- By index:

my_list[index] = new_value # e.g., my_list = 20

- By slice:

my_list[start:end] = [new_values] # e.g., my_list[1:3] =

#### Removing Elements
- By value:

my_list.remove(value) # removes first occurrence

- By index:

del my_list[index]

or

my_list.pop(index)

- Remove last element:

my_list.pop()

- Remove elements by condition (list comprehension):

my_list = [x for x in my_list if x != value]

- Clear entire list:

my_list.clear()

the keys in a dictionary should be unique. these unique keys in a dictionary should be so-called immutable objects. Basically, the content of immutable objects cannot be changed after they're created. Strings, booleans, integers and floats are immutable objects, but the list for example is mutable, because you can change its contents after it's created.

```python
# Dictionary of dictionaries
europe = { 'spain': { 'capital':'madrid', 'population':46.77 },
           'france': { 'capital':'paris', 'population':66.03 },
           'germany': { 'capital':'berlin', 'population':80.62 },
           'norway': { 'capital':'oslo', 'population':5.084 } }


# Print out the capital of France
print(europe['france']['capital'])

# Create sub-dictionary data
data = { 'capital':'rome', 'population':59.83 }

# Add data to europe under key 'italy'
europe['italy'] = data

# Print europe
print(europe)
```

```python
import pandas as pd

# Build cars DataFrame
names = ['United States', 'Australia', 'Japan', 'India', 'Russia', 'Morocco', 'Egypt']
dr =  [True, False, False, False, True, True, True]
cpc = [809, 731, 588, 18, 200, 70, 45]
cars_dict = { 'country':names, 'drives_right':dr, 'cars_per_cap':cpc }
cars = pd.DataFrame(cars_dict)
print(cars)

# Definition of row_labels
row_labels = ['US', 'AUS', 'JPN', 'IN', 'RU', 'MOR', 'EG']

# Specify row labels of cars
cars.index = row_labels

# Print cars again
print(cars)
```

Series : you can think of the Series as a 1-dimensional array that can be labeled, just like the DataFrame.

If you want to select the country column but keep the data in a DataFrame, you'll need double square brackets,

```python
# Series
cars['country'] # columna
cars[0:5] # fila

# DataFrame
cars[['country']]

# 2D array

# .loc[] - label based
cars.loc['RU'] # Series
cars.loc[['RU']] # DataFrame
cars.loc[['RU','IN','US']] # toda la fila
cars.loc[['RU','IN','US'],['country','drives_right']] # filas específicas y columnas específicas
cars.loc[:, ['country','drives_right']] # todas las filas, columnas específicas

# .iloc[] - integer position-based
cars.iloc[[1]] # integer position-based
cars.iloc[[1,2,3]]
cars.iloc[[1,2,3],[1,2]]
cars.iloc[:, [1,2]]
```
